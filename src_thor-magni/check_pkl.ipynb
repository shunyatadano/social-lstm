```python
import pickle

# cpklファイルを読み込む関数
def load_cpkl_file(file_path):
    with open(file_path, 'rb') as file:
        data = pickle.load(file)  # ピクル形式でデータを読み込む
    return data

# 中身を表示する関数
def inspect_data(data):
    # データの型と一部の内容を表示
    print(f"データの型: {type(data)}")
    
    # データが辞書やリストの場合、その内容の一部を表示
    if isinstance(data, (list, dict)):
        print("データの一部:")
        if isinstance(data, list):
            print(data[:5])  # リストなら最初の5つの要素を表示
        elif isinstance(data, dict):
            for key in list(data.keys())[:5]:  # 辞書なら最初の5つのキーと値を表示
                print(f"{key}: {data[key]}")
    else:
        # 他のデータ型でもそのまま表示
        print(data)

# ファイルパスを指定（例: 'dataset.cpkl'）
file_path = '/home/eevee/social-lstm/data/test/trajectories_test.cpkl'

# cpklファイルを読み込む
data = load_cpkl_file(file_path)

# データの中身を確認
inspect_data(data)
```

```python
import pandas as pd

# txtファイルを読み込む関数
def load_dataset(file_path):
    # データをスペースかタブで区切られていると仮定して読み込む
    df = pd.read_csv(file_path, delim_whitespace=True, header=None, names=['frame', 'id', 'y', 'x'])
    return df

# 列の型を確認する関数
def check_data_types(df, sample_size=10):
    # サンプル行を表示して型を確認する
    sample_df = df.sample(n=sample_size)  # サンプルサイズ分のデータを抽出
    data_types = sample_df.dtypes  # 各列のデータ型を表示
    return sample_df, data_types

# ファイルパスを指定（例: 'dataset.txt'）
# file_path = '/home/eevee/social-lstm/data/test/biwi/biwi_eth.txt'
file_path = '/home/eevee/social-lstm/data/validation/thor_magni/20241016_Magni_180522_SC1B_R1.txt'

# データセットを読み込む
df = load_dataset(file_path)

# データのサンプルとその型を表示
sample_df, data_types = check_data_types(df)

# サンプルデータとデータ型を表示
print("サンプルデータ:")
print(sample_df)
print("\nデータ型:")
print(data_types)
```
